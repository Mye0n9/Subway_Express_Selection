```python
def get_stations(path):
  df = pd.read_csv(path, encoding = 'cp949')
  stations = list(df['statnNm'])
  stations = [i.replace('\u3000','') for i in stations]
  return stations
```

```python
key = '--' # 각자 바꿔서 사용하기
gmaps = googlemaps.Client(key)
```

```python
# Getting latitude and longitude data
def get_coord_kr(stations):
  dic = dict()
  for i in stations:
    p_i = i+ ' 역'
    tmp = gmaps.geocode(p_i, language = 'ko')
    if len(tmp) >0:
      dic[i] = tmp[0].get("geometry")["location"] # 외국 데이터면 ko 말고 en 써야할듯
    else:
      dic[i] = 0
  return dic
```

```python
# Getting latitude and longitude data
def get_coord_en(stations):
  dic = dict()
  for i in stations:
    p_i = i+ ' station'
    tmp = gmaps.geocode(p_i, language = 'en')
    if len(tmp) >0:
      dic[i] = tmp[0].get("geometry")["location"] # 외국 데이터면 ko 말고 en 써야할듯
    else:
      dic[i] = 0
  return dic
```

# **시설**
교통(airport, bus_station, taxi_stand) <br>가게(store, cafe, restaurant) <br> 공공,의료시설(police, pharmacy, school) <br> 다중이용시설(amusement_park, art_gallery, campground, casino, church, department_store, museum, shopping_mall, stadium, zoo)

```python
# Type Category
transport = ['airport','bus_station','taxi_stand']
store = ['store', 'cafe', 'restaurant']
public_med_facility = ['fire_station', 'hospital', 'library', 'park', 'police', 'pharmacy', 'school']
multi_facility = ['amusement_park', 'art_gallery', 'campground', 'casino', 'church',
             'department_store', 'museum', 'shopping_mall', 'stadium', 'zoo']
```

```python
def category_counter(coord, facility, stations):
  # URL generator
  urls = []
  for f_type in facility: # 시설 변경 시 수정
    for i in stations:
      # 현재 : Raidus = 200 (200m 반경)
      # Type: Cafe
      if coord[i] == 0:
        pass
      else:
        url_tmp = 'https://maps.googleapis.com/maps/api/place/nearbysearch/json?location='+str(coord[i]['lat'])+'%2C'+str(coord[i]['lng'])+'&radius=100&type='+f_type+'&key='+key
        urls.append(url_tmp)

  # URL to num of facilities
  num_stations = len(stations)
  result = [0]*num_stations
  for i,url in enumerate(urls):
    response = requests.get(url)
    result[i%num_stations] += len(response.json()['results'])
  return result
```

```python
in_coord = seoul_four_coord
in_station = seoul_four

t_result = category_counter(in_coord, transport, in_station)
s_result = category_counter(in_coord, store, in_station)
pm_facility_result = category_counter(in_coord, public_med_facility, in_station)
m_result = category_counter(in_coord, multi_facility, in_station)
```

```python
def df_out(station, transport, store, public, multi):
  data = {
      'statnNm': station,
      'transport': transport,
      'store': store,
      'public_med_facility': public,
      'multi_facility': multi
  }

  # Creating the Pandas DataFrame
  df = pd.DataFrame(data)

  # Displaying the DataFrame
  return df
```

```python
out = df_out(in_station,t_result,s_result,pm_facility_result,m_result)
out
```

```python
df = pd.read_csv(csv_path_8, encoding = 'cp949')
df['statnNm'] = df['statnNm'].str.replace('\u3000', '')
merge_outer = pd.merge(df,out, how='outer',on='statnNm')
merge_outer.to_csv('final_seoulline4.csv')
```
